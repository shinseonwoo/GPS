#### VIF> 10은  i 번째 독립변수 없이도 충분히 반응변수(y)를 잘 설명할 수 있으므로, i 번째 독립변수는 없어도 된다는 것을 의미합니다.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def X_filter_multicollinearity(X, thresh=10.0):
    variables = list(range(X.shape[1]))
    dropped = True
    
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(X.iloc[:, variables].values, i) for i in range(X.iloc[:, variables].shape[1])]
        
        for j in variables:
            print(X.columns[j], ':', vif[variables.index(j)])
            
        maxloc = vif.index(max(vif))

        if max(vif) > thresh:
            print('==> [Dropped variable] : ' + X.iloc[:, variables].columns[maxloc])
            del variables[maxloc]    

            if len(variables) > 1:
                dropped = True
                
        print('-'*50)
        print(' ')

    print('[Remaining variables] :')
    print(X.columns[variables])

    return variables
```

```python
X_remained_idx = X_filter_multicollinearity(X)
```

```python
from sklearn.preprocessing import StandardScaler
df_simple = x_train.drop(columns=['Director_max_award']).copy() 
for c in x_train.columns.drop(['Director_max_award']):
    df_simple[c].fillna(df_simple[c].median(), inplace=True)
    
df_log = x_train.drop(columns=['Director_max_award']).copy() 
for c in x_train.columns.drop(['Director_max_award']):
    df_log[c].fillna(df_log[c].median(), inplace=True)
for c in col_lg:
    df_log[c] = np.log1p(df_log[c])
    
# Take the log1p(x) for all columns
df_alllog = x_train.drop(columns=['Mean_Length']).copy() 
for c in x_train.columns.drop(['Mean_Length']):
    df_alllog[c].fillna(df_alllog[c].median(), inplace=True)
    df_alllog[c] = np.log1p(df_alllog[c])
    

# Scale the features
scaler = StandardScaler()
X_s = scaler.fit_transform(df_simple.values)
X_l = scaler.fit_transform(df_log.values)
X_al = scaler.fit_transform(df_alllog.values)
```
