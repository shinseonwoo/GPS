```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from requests import get
from fake_useragent import UserAgent

from time import sleep
from random import randint
```

```python
ua=UserAgent(verify_ssl=False)
headers={'User-Agent':ua.random}
```

```python
title=[]
release_years=[]
ages=[]
runtimes=[]
genres=[]
ratings=[]
summarys=[]
votes=[]

pages=np.arange(1,7)

for page in pages:
    page=requests.get('https://www.imdb.com/search/keyword/?keywords=netflix-original&ref_=kw_nxt&sort=moviemeter,asc&mode=detail&page='+str(page),headers=headers)
        
    soup=BeautifulSoup(page.text,'html.parser')
    movie_div=soup.find_all('div',class_='lister-item mode-detail')
    
    sleep(randint(2,10))
    
    for container in movie_div:
        #제목
        name=container.h3.a.text
        title.append(name)
        #개봉연도
        year=container.h3.find('span',class_='lister-item-year text-muted unbold').text
        release_years.append(year)
        #연령 제한
        age=container.p.find('span',class_='certificate')if container.p.find('span',class_='certificate')else''
        ages.append(age)
        #상영시간
        runtime=container.p.find('span',class_='runtime')if container.p.find('span',class_='runtime')else''
        runtimes.append(runtime)
        #장르
        genre=container.p.find('span',class_='genre').text
        genres.append(genre)
        #평점
        rating=container.strong
        ratings.append(rating)
        #요약
        summary=container.find('p',class_='').text
        summarys.append(summary)
        #투표수
        vote=container.find('span',attrs={'name':'nv'})
        votes.append(vote)
        
```

```python
imdb=pd.DataFrame({
    'titles':title,
    'release_years':release_years,
    'ages':ages,
    'runtimes':runtimes,
    'genres':genres,
    'ratings':ratings,
    'summarys':summarys,
    'votes':votes
})
```
