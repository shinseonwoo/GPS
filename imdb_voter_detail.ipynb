```python
import pandas as pd
import numpy as np
import random, time
import urllib
from urllib import parse
from urllib.request import urlopen
from bs4 import BeautifulSoup
import openpyxl
import requests
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from IPython.display import display
```

```python
def get_rating_detail(classname):
    product_reviews=[]
    reviews=driver.find_elements_by_class_name(classname)[::]
    for element in reviews:
        product_reviews.append(element.text)
    try:
        result=product_reviews
    except IndexError:
        result='IndexError'
    return result
```

```python
def NextPage():
    try:
        driver.find_element_by_xpath('//*[@id="main"]/div/div[2]/table/tbody/tr[1]/td[2]/a').click()
        
    except NoSuchElementException:
        pass
```

```python
def LastPage():
    try:
        driver.find_element_by_xpath('//*[@id="title-overview-widget"]/div[1]/div[2]/div/div[1]/ div[1]/a').click()
    except  NoSuchElementException:
        pass  
```

```python
data =pd.read_csv('D:/python/Netflix/netflix_originals_합본_2.csv',encoding='ANSI')
title=data.Title.values.tolist()
print(title[:5])
```

```python
title_path = []
for i in title:
    split_title = i.split()
    sum_title = "+".join(split_title)
    title_path.append(sum_title)
title_path[:5]
```

```python
url_list=[]
for i in title_path:
    url="https://www.imdb.com/find?q="+str(i)+ "&ref_=nv_sr_sm"
    url_list.append(url)
```

```python
options = webdriver.ChromeOptions()
options.headless = True # webdriver.ChromeOptions.headless = True (크롬 백그라운드)
options.add_argument('window-size=1920x1080')

driver = webdriver.Chrome(r'C:\Users\seonwoo\Anaconda\chromedriver.exe', options=options)
url_list[::]
get_rating_detail('smallcell')#bigcell
```

```python
page = url_list[:591]    #값 바꿀것
new_data = []


for i in page:
    driver.implicitly_wait(1)
    driver.get(i)
    driver.implicitly_wait(1)
    NextPage()
    driver.implicitly_wait(1)
    LastPage()
    driver.implicitly_wait(1)
    datalist = get_rating_detail('smallcell')
    driver.implicitly_wait(2)
    new_data.append(datalist)
    #driver.get_screenshot_as_file(str(j) + '_screen.png')
    #j += 1
new_data[:5]  # 35
```
